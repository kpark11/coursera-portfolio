## Problems

Now write and execute SQL queries to solve assignment problems

### Problem 1

##### Find the total number of crimes recorded in the CRIME table.


```python
cur.execute('''SELECT COUNT(*) FROM ChicagoCrimeData''')
output_all = cur.fetchall()
for row_all in output_all:
  print(row_all)
```

### Problem 2

##### List community areas with per capita income less than 11000.


```python
cur.execute('''SELECT * FROM ChicagoCensusData WHERE PER_CAPITA_INCOME<11000''')
output_all = cur.fetchall()
for row_all in output_all:
    print(row_all)
```

### Problem 3

##### List all case numbers for crimes  involving minors?(children are not considered minors for the purposes of crime analysis)


```python
cur.execute('''SELECT * FROM ChicagoCrimeData WHERE DESCRIPTION LIKE "%CHILD%"''')
output_all = cur.fetchall()
for row_all in output_all:
    print(row_all)
```

### Problem 4

##### List all kidnapping crimes involving a child?


```python
cur.execute('''SELECT * FROM ChicagoCrimeData WHERE DESCRIPTION LIKE "%ABDUCTION%"''')
output_all = cur.fetchall()
for row_all in output_all:
    print(row_all)
```

### Problem 5

##### What kinds of crimes were recorded at schools?


```python
cur.execute('''SELECT PRIMARY_TYPE FROM ChicagoCrimeData WHERE LOCATION_DESCRIPTION LIKE "%SCHOOL%"''')
output_all = cur.fetchall()
for row_all in output_all:
    print(row_all)
```

### Problem 6

##### List the average safety score for each type of school.


```python
cur.execute('''RETRIEVE * FROM ChicagoPublicSchool''')
output_all = cur.fetchall()
for row_all in output_all:
    print(row_all)
```

```python
cur.execute('''SELECT AVG(SAFETY_SCORE) FROM ChicagoPublicSchool GROUP BY "ES%"''')
output_all = cur.fetchall()
for row_all in output_all:
    print(row_all)
cur.execute('''SELECT AVG(SAFETY_SCORE) FROM ChicagoPublicSchool GROUP BY "MS%"''')
output_all = cur.fetchall()
for row_all in output_all:
    print(row_all)
cur.execute('''SELECT AVG(SAFETY_SCORE) FROM ChicagoPublicSchool GROUP BY "HS"''')
output_all = cur.fetchall()
for row_all in output_all:
    print(row_all)
```

### Problem 7

##### List 5 community areas with highest % of households below poverty line


```python
cur.execute('''SELECT * FROM ChicagoCensusData ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY DESC''')
output_all = cur.fetchall()
for row_all in output_all:
    print(row_all)
```

### Problem 8

##### Which community area is most crime prone?


```python
cur.execute('''SELECT community_area_number FROM ChicagoCrimeData GROUP BY community_area_number ORDER BY COUNT(*) DESC''')
output_all = cur.fetchall()
for row_all in output_all:
    print(row_all)
```

```python
cur.execute('''SELECT * FROM ChicagoCensusData WHERE community_area_number=25''')
output_all = cur.fetchall()
for row_all in output_all:
    print(row_all)
```

### Problem 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
cur.execute('''SELECT (SELECT COMMUNITY_AREA_NAME FROM ChicagoCensusData ORDER BY hardship_index DESC) FROM ChicagoCrimeData''')
output_all = cur.fetchall()
for row_all in output_all:
    print(row_all)
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
cur.execute('''SELECT (SELECT COMMUNITY_AREA_NAME FROM ChicagoCrimeData GROUP BY community_area_number ORDER BY COUNT(*) DESC) FROM ChicagoCensusData''')
output_all = cur.fetchall()
for row_all in output_all:
    print(row_all)
```
